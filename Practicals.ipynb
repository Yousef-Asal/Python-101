### 7.5.1 Mutable versus Immutable

```python
#Q1
my_list = ['dog', 'cat', 'ant']
print(id(my_list), my_list)
my_list[2] = 'cow'
print(id(my_list), my_list)

#Q2
def change_item(lst, idx, new_val):
    if type(lst) is list or type(lst) is tuple:
        if idx >= len(lst):
            print('Can not change the value, Index is out of bounds')
        else:
            lst[idx] = new_val
    elif type(lst) is dict:
        k = list(lst.keys())
        if idx >= len(k):
            print('Can not change the value, Index is out of bounds')
        else:
            lst[k[idx]] = new_val

new_lst = [1,2,15,4]
print(id(new_lst), new_lst)
change_item(new_lst, 2, 3)
print(id(new_lst), new_lst)

#Q3 
my_dict = {'first': 1, 'second': 2, 'third': 5}
print(id(my_dict), my_dict)
change_item(my_dict, 2, 3)
print(id(my_dict), my_dict)

#Q4 
# str0 = 'Hello World!!'
# print(id(str0), str0)
# str0[3] = 'k'
# print(id(str0), str0)
# This code will raise an error because str is immutable so we can't change its content after its created

#Q5 
# my_tuple = (1,2,3,4)
# change_item(my_tuple, 2, 3)
# This will also raise an error because tuples are immutable. 
```

### 7.5.2 More on Functions

```python
def rect(*args, **kwargs):
    if 'round_factor' in kwargs:
        r_factor = kwargs['round_factor']
    else:
        r_factor = 2

    if len(args) == 1:
        print('This is a Square')
        circumference = round(4*args[0], r_factor)
        area = round(args[0] ** 2, r_factor) 
    elif len(args) == 2:
        print('This is a Rectangle')
        circumference = round(2 * (args[0]+args[1]), r_factor)
        area = round(args[0] * args[1], r_factor)

    return circumference, area

print(rect(3.25, round_factor= 3))
print(rect(3.52, 5.22))
```

### 7.5.3 Decorators

```python
import time

def my_func(func):
    def internal(N):
        start_time = time.time()
        func(N)
        end_time = time.time()
        dur = end_time - start_time
        print(f'The function took {dur} seconds')
    
    return internal

@my_func
def foo(N):
    y = 0
    for i in range(N):
        y += 1

foo(1000)
```

## Ch8. Basics of Classes

### 8.4.1 Classes

```python
#Q1 
class shape:
    def __init__(self, s: str):
        self.shape = s
    
    def print_description(self):
        print(f"This shape is a {self.shape}")

obj0 = shape('square')
obj0.print_description()

#Q2,Q3
class circle(shape):
    def __init__(self, s, r):
        super().__init__(s)
        self.radius = r
    
    def diameter(self):
        self.d = self.radius * 2
    
    def circumference(self):
        self.c = self.radius* 2 * 3.14
    
    def area(self):
        self.a = 3.14 * (self.radius ** 2)

    def calculate(self):
        self.diameter()
        self.circumference()
        self.area()
        return self.d, self.c, self.a
    
    def print_description(self):
        print(f"This shape is a {self.shape}")
        print(f"Circle Radius is: {self.radius} \n- Diameter: {self.d} \n- Circumference: {self.c} \n- Area: {self.a}")

obj1 = circle('circle', 1)
# Note that if we tried printing the description before calcutaling it will give us an errer
# because we didn't inialize the values so they don't exist unless we call calculate() 
# obj1.print_description()
d, c, a = obj1.calculate()
obj1.print_description()

    
#Q4
class rect(shape):
    def __init__(self, s, h, w):
        super().__init__(s)
        self.height = h
        self.width = w
        self.c = 0
        self.a = 0
    
    def circumference(self):
        self.c = 2 * (self.height + self.width)
    
    def area(self):
        self.a = self.height * self.width

    def calculate(self):
        self.circumference()
        self.area()
        return self.c, self.a
    
    def print_description(self):
        print(f"This shape is a {self.shape}")
        print(f"Rect Height is: {self.height} \n- Width: {self.width} \n- Circumference: {self.c} \n- Area: {self.a}")

obj2 = rect('Rectangle', 3, 2)
obj2.print_description()
c, a = obj2.calculate()
obj2.print_description()
# Here since we initialized the circumference and area we don't get an error if we
# ran print_description() before calculate()
```

### 11.5.2 Class Strings

```python
class rectangle():
    def __init__(self,*args):
        if len(args) == 1:
            self.shape = 'Square'
            self.height = args[0]
            self.width = args[0]
        else:
            self.shape = 'Rectangle'            
            self.height = args[0]
            self.width = args[1]

    def calculate_circumference(self):
        self.circ = 2*(self.height+self.width)
        return self.circ

    def calculate_area(self):
        self.area = self.height*self.width
        return self.area
    
    def __str__(self):
        self.calculate_circumference()
        self.calculate_area()
        return f"This is a {self.shape}\n - Height = {self.height}\n - Width = {self.width}\n - Circumference = {self.circ}\n - Area = {self.area}"
    
    def str(self):
        self.calculate_circumference()
        self.calculate_area()
        return f"This is a {self.shape}\n - Height = {self.height}\n - Width = {self.width}\n - Circumference = {self.circ}\n - Area = {self.area}"
    
sqr = rectangle(5)
rect = rectangle(4,8)
s1 = str(sqr)
s2 = str(rect)
print(s1,sqr,sep='\n')
print(s2,rect,sep='\n')

# the str(self) wasn't used in any of the calls since str(sqr) and str(rect) internally calls __str__(self) method
# Also print internally calls __str__(self) method so we never use str(self)
# To use it we have to run it like this:
# print(sqr.str()) but that's extra steps since it's the same as __str__(self) so the str(self) method is redundent
```

### 11.5.3 Class With

```python
import time

class timer:
    def __init__(self,name):
        self.name = name

    def iterate(self,n):
        self.n_iter = n
        y = 0
        for i in range(self.n_iter):
            y += 1

    def __enter__(self):
        self.start_time = time.time()
        return self

    def __exit__(self,type,value,traceback):
        self.dur = time.time() - self.start_time
        print(f"The iterator {self.name} took {self.dur} seconds to iterate {self.n_iter} iterations")
    
with timer('Thousander') as itr:
    itr.iterate(1000)

with timer('TenThousander') as itr:
    itr.iterate(10000)

with timer('HunderedThousander') as itr:
    itr.iterate(100000)
```
